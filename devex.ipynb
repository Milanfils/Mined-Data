```python
#import libraries

import pandas as pd
import numpy as np
import time
import math
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver import Firefox, Chrome
from selenium.webdriver import ActionChains
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

options = Options()
#options.headless = True
options.add_argument("--no-sandbox")
options.add_argument("--disable-gpu")
options.add_argument("--window-size=1920x1080")
```

```python
#start web browser
s = Service(ChromeDriverManager().install())
driver = Chrome(service=s, options=options)
```

```python
driver.get("https://www.devex.com/organizations/search")
time.sleep(8)
```

```python
#calaculates total number of pages to iterate
npg = driver.find_element(By.CSS_SELECTOR, "span[ng-show='pagination']").text.strip().split("of ")[1].replace(",", "")
npg = int(math.ceil(int(npg)/20))
```

```python
#lists for storing tab1 and tab2 data
data1 = []
data2 = []
```

```python
#for pg in np.arange(1, npg+1):
count1 = 45
#iterates through all pages
for pg in np.arange(45, npg+1):
    print('count1 is '+ str(count1))
    count1 += 1
    url = "https://www.devex.com/organizations/search?page%5Bnumber%5D=" + str(pg)
    driver.get(url)
    time.sleep(15)
    
    links = driver.find_elements(By.CSS_SELECTOR, "ul.result-summary-list > li div.info-container > a")
    links = [i.get_attribute("href") for i in links]
    count2 = 1
    #iterates through each organization to extract data
    for add in links:
        print('count2 is '+ str(count2))
        count2 += 1
        driver.get(add)
        time.sleep(15)
    
        try:
            company_name = driver.find_element(By.CSS_SELECTOR, "h1.emphasized.org-name").text.strip()
        except:
            company_name = ""

        try:
            company_logo = driver.find_element(By.CSS_SELECTOR, "div.org-container > div > img").get_attribute("src")
        except:
            company_logo = ""

        try:
            company_description = driver.find_element(By.CSS_SELECTOR, "div#description-container > p").text
        except:
            company_description = ""

        try:
            organization_type = driver.find_element(By.XPATH, "//span[contains(text(), 'Organization Type')]/./following-sibling::strong").text
        except:
            organization_type = ""

        try:
            staff = driver.find_element(By.XPATH, "//span[contains(text(), 'Staff')]/./following-sibling::strong").text
        except:
            staff = ""

        try:
            development_budget = driver.find_element(By.XPATH, "//span[contains(text(), 'Development Budget')]/./following-sibling::strong").text
        except:
            development_budget = ""

        try:
            headquarters = driver.find_element(By.XPATH, "//span[contains(text(), 'Headquarters')]/./following-sibling::strong").text
        except:
            headquarters = ""

        try:
            founded = driver.find_element(By.XPATH, "//span[contains(text(), 'Founded')]/./following-sibling::strong").text
        except:
            founded = ""

        try:
            organization_website = driver.find_element(By.CSS_SELECTOR, "small.organization-website > a").get_attribute("href")
        except:
            organization_website = ""

        try:
            sec_tx = driver.find_elements(By.XPATH, "//h2[contains(text(), 'Sectors')]/./following-sibling::ul/li")
            sectors = [i.text for i in sec_tx]    
        except:
            sectors = ""

        try:
            fun_tx = driver.find_elements(By.XPATH, "//h2[contains(text(), 'Funders')]/./following-sibling::ul/li")
            funders = [i.text for i in fun_tx]  
        except:
            funders = ""

        try:
            con_tx = driver.find_elements(By.XPATH, "//h2[contains(text(), 'Countries')]/./following-sibling::ul/li")
            countries = [i.text for i in con_tx]
        except:
            countries = ""

        try:
            skill_tx = driver.find_elements(By.XPATH, "//h2[contains(text(), 'Skills')]/./following-sibling::ul/li")
            skills = [i.text for i in skill_tx]
        except:
            skills = ""

        dict1 = {"Company Name": company_name, "Company Logo": company_logo, "Company Description": company_description,
                "Organization Type": organization_type, "Staff": staff, "Development Budget": development_budget,
                "Head Quaters": headquarters, "Founded": founded, "Organization Website": organization_website,
                "Sectors": sectors, "Funders": funders, "Countries": countries, "Skills": skills}
        data1.append(dict1)        

        contracts = driver.find_elements(By.XPATH, "//h3[contains(text(), 'Contract Awards ')]/./following-sibling::div/div")

        if len(contracts) == 0:
            contract_name = ""
            contract_fundier = ""
            dict2 = {"Company Name": company_name,
                    "Contract Name": contract_name,
                    "Contract Fundier": contract_fundier}
            data2.append(dict2)
        else:
            for con in contracts:

                try:
                    contract_name = con.find_element(By.CSS_SELECTOR, "h4").text
                except:
                    contract_name = ""

                try:
                    contract_fundier = con.find_element(By.CSS_SELECTOR, "div.text-gray-li").text
                except:
                    contract_fundier = ""

                dict2 = {"Company Name": company_name,
                        "Contract Name": contract_name,
                        "Contract Fundier": contract_fundier}
                data2.append(dict2)

           
```

```python
#turns data to a dataframe                
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2) 
```
